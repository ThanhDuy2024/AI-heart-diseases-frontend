# **Framingham-Heart-Study**
## Trong dữ liệu này ta sẽ tiến hành huấn luyện một mô hình chuẩn đoán bệnh nhên có mắc bệnh tim mạch hay không

# **Một số đặc trưng cần chú ý**
- **sex** : Trong cột dữ liệu là male với 0 là nam, 1 là nữ.
- **age** : Tuổi của bệnh nhân
- **education** : Giáo dục với **với 1 là đang học trung học**, **2 là đã tốt nghiệp trung học**, **3 là Sinh viên**, **4 đã tốt nghiệp**
- **currentSmoker**: Tình trạng hút thuốc lá **(0 là ko hút, 1 là có hút)**
- **cigsPerDay**: Số lần hút thuốc trong ngày
- **BPmeds**: Có sử dụng thuốc giảm huyết áp trong quá trình kiểm tra hay không ? **(0 là không có, 1 là có)**
- **prevalentStroke**: Tỉ lệ mắc bệnh đột quỵ **(0 = Không là không mắc bệnh)**
- **prevalentHyp**: Tăng huyết áp hiện có **(0 là không tăng huyết áp, 1 là có tăng huyết áp)**
- **diabetes**: Có mắc bệnh tiểu đường hay không ? **(0 là không mắc, 1 là có mắc)**
- **totChol**: Tổng số cholesterol (mg/dL)
- **sysBP**: Huyết áp tâm thu (mmHg)
- **diaBP**: Huyết áp tâm trương (mmHg)
- **BMI**: Chỉ số khối cơ thể, cân nặng (kg)/chiều cao (m)^2
- **heartRate**: Nhịp tim (số nhịp/phút)
- **Glucose**: Nồng độ đường huyết (mg/dL)
- Và cuối cùng là biến phản hồi:
  + **TenYearCHD**: Nguy cơ mắc bệnh tim mạch vành (CHD) trong 10 năm. (0 là không mắc, 1 là có mắc bệnh)

```python
import pandas as pd
data = pd.read_csv("./data/framingham.csv")
data.head()
```

```python
#Check total columns and rows
data.shape
```

```python
#Check name columns
data.columns
```

```python
#Check missing data and Dtype
data.info();
```

```python
#cigsPerDay is missing data
#education is missing data
#BPMeds is missing data
#totChol is missing data
#BMI is missing data
#heartRate is missing data
#glucose is missing data
data.isnull().sum()
```

```python
#Fix missing data
def impute_median(data):
    return data.fillna(data.median())


data.glucose = data['glucose'].transform(impute_median)
data.education = data['education'].transform(impute_median)
data.heartRate = data['heartRate'].transform(impute_median)
data.totChol = data['totChol'].transform(impute_median)
data.BPMeds = data['BPMeds'].transform(impute_median)

by_currentSmoker = data.groupby(['currentSmoker'])
data.cigsPerDay = by_currentSmoker['cigsPerDay'].transform(impute_median)

by_age = data.groupby(['male','age'])
data.BMI = by_age['BMI'].transform(impute_median)
data.isnull().sum()
```

```python
#Explore descript data
data.describe()
```

# Tổng quan về dữ liệu 
- age : Max là 70, min là 32, mean là 49
- cigsPerDay : Max là 70 điếu thuốc, min là  0, mean là 9.08
- totChol : Max là 696, min là 44.32 , mean là 236
- sysBP : Max là 295, min là 83.5 , mean là 132
- diaBP : Max là 142, min là 48 , mean là 82.9
- BMI : Max là 56.8, min là 15.54 , mean là 25.7
- heartRate : Max là 143, min là 44 , mean là 75.8
- glucose : Max là 394, min là 40 , mean là 81.6

```python
#Split 80% train and 20% test
from sklearn.model_selection import train_test_split
x = data.drop("TenYearCHD", axis=1)
y = data["TenYearCHD"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Scale data
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_train = scale.fit_transform(x_train)
x_test = scale.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Training model and report
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
model = RandomForestClassifier()
model = model.fit(x_train, y_train)
predict = model.predict(x_test)
report = classification_report(y_test, predict)
print(report)
```

```python
percent = model.predict_proba(x_test)
print(percent)
```

```python
patient_df = pd.DataFrame([{
    "male": 0,
    "age": 61,
    "education": 3.0,
    "currentSmoker": 1,
    "cigsPerDay": 30.0,
    "BPMeds": 0.0,
    "prevalentStroke": 0,
    "prevalentHyp": 1,
    "diabetes": 0,
    "totChol": 225.0,
    "sysBP": 150.0,
    "diaBP": 95.0,
    "BMI": 28.58,
    "heartRate": 65.0,
    "glucose": 103.0
}])

patient_scaled = scale.transform(patient_df)

prediction = model.predict(patient_scaled)
probability = model.predict_proba(patient_scaled)

print("Dự đoán:", prediction[0])
print("Xác suất mắc bệnh:", probability[0][1] * 100, "%")
```
